# matplotlib基础

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = np.linspace(0, 10, 100)
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x, y) # 传入横坐标对应的值和纵坐标对应的值
```

```python
plt.show() # 展示图像
```

```python
cosy = np.cos(x)
```

```python
cosy.shape
```

```python
siny = y.copy()
```

```python
plt.plot(x, siny) # 两条曲线
plt.plot(x, cosy)
plt.show()
```

```python
plt.plot(x, siny) # 两条曲线
plt.plot(x, cosy, color="red") # 设置曲线颜色
plt.show()
```

```python
plt.plot(x, siny) # 两条曲线
plt.plot(x, cosy, color="red", linestyle="--") # 设置曲线颜色, 设置曲线形状
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.xlim(-5, 15) # 设置x坐标轴范围
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.xlim(-5, 15) # 设置x坐标轴范围
plt.ylim(0, 1.5) # 设置y坐标轴范围
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.axis([-1, 11, -2, 2]) # 设置坐标轴范围
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.xlabel("x axis")
plt.ylabel("y axis") # 设置坐标轴的标签
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)") # label设置图示
plt.xlabel("x axis")
plt.ylabel("y axis") # 设置坐标轴的标签
plt.legend() # 显示图示
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)") # label设置图示
plt.xlabel("x axis")
plt.ylabel("y axis") # 设置坐标轴的标签
plt.legend() # 显示图示
plt.title("Welcome to the ML world") # 添加标题
plt.show()
```

## Scatter Plot(散点图)

```python
plt.scatter(x, siny)
plt.show()
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color="red") # 散点图通常绘制二维特征
plt.show()
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)
plt.scatter(x, y, alpha=0.1) # alpha透明度
plt.show()
```

```python
t = np.linspace(-10, 10, 1000)
x = 16 * np.power(np.sin(t), 3)
y = 13 * np.cos(t) - 5 * np.cos(2 * t) - 2 * np.cos(3 * t) - np.cos(4 * t)
plt.plot(x, y, color='red', linewidth=3, label='love')
plt.xlabel('thw')
plt.ylabel('glf')
plt.xlim(-20, 20)
plt.title("Love You")
plt.legend()
plt.show()
```
